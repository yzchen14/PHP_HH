```python
import pandas as pd
import numpy as np
```

```python
data =pd.read_excel("sampledatainsurance.xlsx")
```

```python
data
```

```python
table1 = pd.pivot_table(data,  index = "Expiry", columns = "Location", values = "InsuredValue", aggfunc=np.sum, dropna=True)
```

```python
table1
```

```python
import plotly.io as pio
import plotly.graph_objects as go
from plotly.subplots import make_subplots



pio.templates.default = "plotly_white"


fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Scatter(x=table1.index, y=table1["Urban"], name="yaxis data", mode = "markers"#mode='lines+markers'
               , marker = dict(size = 5, color = "rgb(255,0,0)"),
               
              
              ), secondary_y=False,
)

fig.add_trace(
    go.Bar(x=table1.index, y=table1["Urban"] * 50, name="yaxis data", marker_color = "#cceeff"),
    secondary_y=True,
)

#fig =  px.line(x=table1.index, y=table1["Urban"])
#fig.add_bar(x=table1.index, y=table1["Urban"] * 5, secondary_y=True)
fig.show()
```
